```python
import pandas as pd

ohlc = pd.read_csv('../data/ohlc_btc.csv')
ohlc.info(memory_usage='deep')
```

### I. Just an example to select values from df

```python
#Possible method to obtain only values from 2018
ohlc_ = ohlc.copy()
ohlc_['str_ts'] = ohlc_.opentime.apply(lambda x: pd.Timestamp(x, unit='s')).dt.strftime('%Y-%m-%d')

ohlc_[ohlc_.str_ts.apply(lambda x: x.startswith('2018-01'))]
```

```python
# Create empty dict
df_by_months = {}

# Obtain the months by year
keys_collection = ohlc_.opentime.apply(
    lambda x: pd.Timestamp(x, unit='s')).dt.strftime('%Y-%m').unique()

# Iterate through the keys collection == each month of each year
for key in keys_collection:
    df_by_months[key] = ohlc_[ohlc_.str_ts.apply(lambda x: x.startswith(key))]
```

```python
df_by_months['2018-01']
```

### II. Back to the exercice

```python
ohlc.index = [pd.Timestamp(ix, unit='s') for ix in ohlc.opentime]
ohlc.drop(columns='opentime', inplace=True)
ohlc.head()
```

### III. Plot the close of BTC daily candles

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(17, 8))
plt.title('BTCUSDT on binance chart')

plt.plot(ohlc.close)
plt.xlabel('timestamp')
plt.ylabel('USDT');
```

### IV. Describe your data

```python
# Use the index to obtain 2018 only
OHLC_RANGE = ohlc.loc['2018-01-01': '2018-12-31']

# Store all stats in one frame
stats18 = pd.concat([
    (OHLC_RANGE.mean()).astype(int),
    (OHLC_RANGE.median()).astype(int),
    (OHLC_RANGE.min()).astype(int),
    (OHLC_RANGE.max()).astype(int),
    OHLC_RANGE.quantile(0.25).astype(int),
    OHLC_RANGE.quantile(0.75).astype(int)
], axis=1).T
stats18.index = ['mean18', 'median18', 'min18', 'max18', 'qt25_18', 'qt75_18']
stats18
```

```python
# Same thing from 2019

OHLC_RANGE = ohlc.loc['2019-01-01': '2019-12-31']

stats19 = pd.concat([
    OHLC_RANGE.mean().astype(int),
    OHLC_RANGE.median().astype(int),
    OHLC_RANGE.min().astype(int),
    OHLC_RANGE.max().astype(int),
    OHLC_RANGE.quantile(0.25).astype(int),
    OHLC_RANGE.quantile(0.75).astype(int)
], axis=1).T
stats19.index = ['mean19', 'median19', 'min19', 'max19', 'qt25_19', 'qt75_19']
stats19
```

```python
# Store all stats...
stats = pd.concat([stats18, stats19])

# And plot it with pandas (this is not matplotlib here)
stats.iloc[:, :4].plot.bar(
    figsize=(10, 5), 
    title='BTC-USDT: Descriptive statistics for year 2018 and 2019',
);
```

### V. Calcul Simple Moving Average for timeframes 10, 20, 50

```python
ohlc['SMA_10'] = ohlc.close.rolling(10).mean()
ohlc['SMA_20'] = ohlc.close.rolling(20).mean()
ohlc['SMA_50'] = ohlc.close.rolling(50).mean()
```

```python
# Count number of NA generate by the rolling function
ohlc.isna().sum()
```

```python
# Display this data with matplotlib

# Params
plt.figure(figsize=(17, 8))
plt.title('BTCUSDT on binance chart with SMA')

# Plot
plt.plot(ohlc.close, label='BTC-USDT', color='gray')
plt.plot(ohlc.SMA_10, label='SMA_10', color='red')
plt.plot(ohlc.SMA_20, label='SMA_20', color='green', linewidth=1.2)
plt.plot(ohlc.SMA_50, label='SMA_50', color='blue', alpha=0.5)

plt.xlabel('timestamp')
plt.ylabel('USDT')
```

### VI. Calcul bollinger bands

```python
# calculate the standard deviation
rstd = ohlc.close.rolling(20).std()

# Apply the math to obtain upper and lower band
upper_band = ohlc.SMA_10 + 2 * rstd
upper_band.name = 'upper_band'
lower_band = ohlc.SMA_10 - 2 * rstd
lower_band.name = 'lower_band'
```

```python
# Plot with pandas again
pd.concat([ohlc.close, upper_band, lower_band], axis=1).plot(
    figsize=(17, 10), 
    title='Bollinger band and BTC-USDT closing candles',
    ylabel='USDT'
);
```

### VII. Triggers for bearish period and bull period

```python
bearish_period = ohlc[(ohlc.SMA_10 < ohlc.SMA_20) & (ohlc.SMA_10 < ohlc.SMA_50)].index
```

```python
bullish_period = ohlc[(ohlc.SMA_10 > ohlc.SMA_20) & (ohlc.SMA_10 > ohlc.SMA_50)].index
```

```python
ohlc['label_period'] = 'no_trends'


ohlc.loc[bearish_period, 'label_period'] = 'bearish'
ohlc.loc[bullish_period, 'label_period'] = 'bullish'
no_trends = ohlc[ohlc.label_period == 'no_trends']
```

```python
ohlc.label_period.value_counts()
```

```python
no_trends
```

```python
plt.figure(figsize=(17, 8))
plt.title('BTCUSDT on binance chart with SMA')

plt.plot(ohlc.close, label='BTC-USDT', color='gray')
plt.bar(bearish_period, ohlc.close.max(), color='red', alpha=0.1)
plt.bar(bullish_period, ohlc.close.max(), color='green', alpha=0.1)
plt.bar(no_trends.index, ohlc.close.max(), color='gray', alpha=0.3)
plt.plot(ohlc.SMA_10, label='SMA_20', color='black', linewidth=1.6)
plt.plot(ohlc.SMA_20, label='SMA_20', color='cyan', linewidth=1.6)
plt.plot(ohlc.SMA_50, label='SMA_50', color='blue')
plt.xlabel('timestamp')
plt.ylabel('USDT')
```

### VII. Linear regression 

```python
from scipy import stats
import numpy as np

X, y = np.array(ohlc.reset_index().index[-30:]), np.array(ohlc.close[-30:])
lr = stats.linregress(X, y)
print(lr)


plt.scatter(X, y)
plt.plot(X, lr[0]*X+lr[1], '--k');
```
