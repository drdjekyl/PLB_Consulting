## Titanic from Kaggle

```python
import pandas as pd

df = pd.read_csv('../../data/titanic.csv')
```

### Informations

```python
df.info()
```

```python
df.drop(columns='PassengerId', inplace=True)
```

### Missing values

```python
import matplotlib.pyplot as plt

plt.bar(
    df.isna().sum().index, 
    df.isna().sum().values
)
plt.xticks(rotation=90);
```

### Plot some informations about features

```python
for col in ['Survived', 'Pclass', 'Sex', 'Age', 'Fare', 'Cabin', 'Embarked']:
    # CREATE STATS
    distrib = pd.Series(df.value_counts(col).sort_index(), name=col)
    
    # RECORDS
    distrib.to_csv(f'../../images/titanic/distrib_{col}.png')
    
    #display(distrib.head(3))
    
    # PLOT
    plt.figure(figsize=(15, 6))
    plt.bar(distrib.index, distrib.values)
    plt.xticks(rotation=90)
    plt.title(f'{col} distribution')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.bar(df.Age.value_counts().index, df.Age.value_counts().values)
plt.title('Age distribution');
```
